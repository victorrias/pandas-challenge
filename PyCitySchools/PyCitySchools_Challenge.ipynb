```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)

school_data_to_load = "/Users/victoriasandoval/UCI-VIRT-DATA-PT-03-2022-U-LOL/02-Homework/04-Pandas/Instructions/PyCitySchools/Resources/schools_complete.csv"
student_data_to_load = "/Users/victoriasandoval/UCI-VIRT-DATA-PT-03-2022-U-LOL/02-Homework/04-Pandas/Instructions/PyCitySchools/Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
school_data_complete.head()
```

```python
total_schools = school_data_complete["school_name"].nunique()
total_schools
```

```python
total_students = school_data_complete["Student ID"].nunique()
total_students
```

```python
total_budget = school_data["budget"].sum()
total_budget
```

```python
avg_math_score = school_data_complete["math_score"].mean()
avg_math_score
```

```python
avg_reading_score = school_data_complete["reading_score"].mean()
avg_reading_score
```

```python
students_passing_math = school_data_complete.loc[school_data_complete['math_score'] >=70].count()["Student ID"]/total_students*100
students_passing_math
```

```python
students_passing_reading = school_data_complete.loc[school_data_complete['reading_score'] >=70].count()["Student ID"]/total_students*100
students_passing_reading
```

```python
Overall_passing_score = school_data_complete.loc[(school_data_complete['math_score'] >= 70) &
 (school_data_complete['reading_score'] >= 70)].count()["Student ID"]/total_students*100
Overall_passing_score
```

```python
District_Summary = pd.DataFrame(
                    {"Total Number of Schools": [total_schools],
                    "Total Number of Students": [total_students],
                    "Total Budget" : [total_budget],
                    "Average Math Score": [avg_math_score],
                    "Average Reading Score" : [avg_reading_score],
                    "% Passing Math" : [students_passing_math],
                    "% Passing Reading" : [students_passing_reading],
                    "% Passing Overall" : [Overall_passing_score]})

District_Summary
```
